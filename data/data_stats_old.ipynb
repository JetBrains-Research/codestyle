```python
for file_id in range(data_count):
    df = pd.read_csv(data_file.format(file_id))
    total = len(df)
    print("Total number of methods: {}".format(total))
    created = df['methodBeforeId'].value_counts()[-1]
    deleted = df['methodAfterId'].value_counts()[-1]
    changed = total - created - deleted
    print("Created: {}, deleted: {}, changed: {}".format(created, deleted, changed))
    count += total
    creations += created
    deletions += deleted
    changes += changed
    for name, cnt in df['authorName'].value_counts().iteritems():
        author_counts[name] += cnt
        
    for name, cnt in df['authorEmail'].value_counts().iteritems():
        email_counts[name] += cnt

    del df

print("---------------")
print("Total number of methods: {}".format(count))
print("Created: {}, deleted: {}, changed: {}".format(creations, deletions, changes))
print("---------------")
```

```python
lengths_before = Counter()
lengths_after = Counter()
lengths_changes = Counter()

for file_id in range(data_count):
    print("Processing #{}".format(file_id + 1))
    df = pd.read_csv(data_file.format(file_id))
    for index, row in df.iterrows():
        if row['methodBeforeId'] != -1:
            lengths_before[row['pathsCountBefore']] += 1
        if row['methodAfterId'] != -1:
            lengths_after[row['pathsCountAfter']] += 1
        if row['methodBeforeId'] != -1 and row['methodAfterId'] != -1:
            lengths_changes[row['pathsCountAfter'] - row['pathsCountBefore']] += 1
    del df
```

```python
def print_counter(cntr):
    for key, value in sorted(cntr.items(), key=lambda it: it[0]):
        print("{};{}".format(key, value))
```

```python
print_counter(lengths_before)
print("--------------------")
print_counter(lengths_after)
print("--------------------")
print_counter(lengths_changes)
```

```python
def read_tokens(token_file):
    with open(token_file, 'r') as f:
        indices = []
        tokens = []
        for line in f:
            if not ',' in line:
                tokens[-1] += line
                continue
            ind, token = line.split(',', 1)
            indices.append(ind)
            tokens.append(token)
        for i in range(len(tokens)):
            tokens[i] = tokens[i][:-1]
        return indices, tokens

indices, tokens = read_tokens('idea-changes/tokens.csv')
id_name, indices = indices[0], indices[1:]
token_name, tokens = tokens[0], tokens[1:]
indices = list(map(int, indices))
```

```python
email2author = {}
author2email = {}
for file_id in range(data_count):
    print("Processing #{}".format(file_id + 1))
    df = pd.read_csv(data_file.format(file_id))
    for index, row in df.iterrows():
        email2author[row['authorEmail']] = set()
        author2email[row['authorName']] = set()
    del df
    

for file_id in range(data_count):
    print("Processing #{}".format(file_id + 1))
    df = pd.read_csv(data_file.format(file_id))
    for index, row in df.iterrows():
        email2author[row['authorEmail']].add(row['authorName'])
        author2email[row['authorName']].add(row['authorEmail'])
    del df
```

```python
component_mark = {}
items = {}
ind = 0
for email in email2author.keys():
    if email == 'no_reply@jetbrains.com':
        continue
    component_mark[(email, 0)] = ind
    items[ind] = [(email, 0)]
    ind += 1
for author in author2email.keys():
    if author == 'unknown' or author == 'no_reply@jetbrains.com':
        continue
    component_mark[(author, 1)] = ind
    items[ind] = [(author, 1)]
    ind += 1
```

```python
def merge_components(ind1, ind2):
    if ind1 == ind2:
        return
    for item in items[ind1]:
        component_mark[item] = ind2
        items[ind2].append(item)
    items[ind1] = []
```

```python
def is_bad(author, email):
    return author == 'unknown' or author == 'no_reply@jetbrains.com' or email == 'no_reply@jetbrains.com'

for author, emails in author2email.items():
    for email in emails:
        if is_bad(author, email):
            continue
        merge_components(component_mark[(author, 1)], component_mark[(email, 0)])
for email, authors in email2author.items():
    for author in authors:
        if is_bad(author, email):
            continue
        merge_components(component_mark[(author, 1)], component_mark[(email, 0)])
```

```python
for component in items.values():
    if component != []:
        total_author = 0
        total_emails = 0
        output = ''
        for val, t in component:
            if t == 0:
                total_emails += email_counts[val]
            else:
                total_author += author_counts[val]
            if output != '':
                output += ';'
            output += val + '-' + str(t)
        output += ';' + str(total_author) + ';' + str(total_emails)
        print(output)
```
